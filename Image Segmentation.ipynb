```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
image = '/kaggle/input/car-detection-and-tracking-dataset/archive/train/images/frame_0000.jpg'
```

```python
img = cv2.imread(image)
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img.shape
```

```python
pixel_vals = img.reshape((-1, 3))
```

```python
pixel_vals = np.float32(pixel_vals)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
```

```python
retvals, labels, centers = cv2.kmeans(pixel_vals, 5, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)
```

```python
segmented_data = centers[labels.flatten()]
```

```python
segmented_data = segmented_data.reshape((img.shape))
```

```python
plt.imshow(segmented_data)
plt.axis('off')
plt.show()
```
