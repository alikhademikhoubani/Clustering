```python
def object_columns(df):
    for column in df:
        if df[column].dtype=='object':
            print(f'{column}:{df[column].unique()}')
```

```python
object_columns(df)
```

```python
one_hot=OneHotEncoder()
```

```python
transformed=one_hot.fit_transform(df[['job']])
```

```python
print(transformed.toarray())
```

```python
print(one_hot.categories_)
```

```python
df[one_hot.categories_[0]]=transformed.toarray()
print(df.head())
```

```python
df=pd.DataFrame(df)
df.head()
```

```python
df.drop('job',axis='columns',inplace=True)
```

```python
df.head()
```

```python
object_columns(df)
```

```python
cols=['marital','education','default','housing','loan','month','day_of_week','poutcome']
```

```python
df=pd.get_dummies(df,columns=cols,drop_first=True)
```

```python
df.head()
```

```python
object_columns(df)
```

```python
df['contact']=df['contact'].replace({'telephone':0,'cellular':1})
```

```python
df.sample(4)
```

```python
object_columns(df)
```

```python
df.info()
```

```python
df.shape
```

```python
df.std().mean()
```

```python
scaler=MinMaxScaler()
```

```python
for column in df:
    print(f'{column}:{df[column].unique()}')
```

```python
cols2=['age','duration','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
```

```python
df[cols2]=scaler.fit_transform(df[cols2])
```

```python
df.head()
```

```python
df.std().mean()
```

```python
k_rng=range(1,20)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit_transform(df)
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.show()
```

```python
km=KMeans(n_clusters=2)
```

```python
y_pred=km.fit_transform(df)
```

```python
y_pred
```

```python
y_predict=[np.argmax(element) for element in y_pred]
```

```python
y_predict[15:20]
```

```python
df['cluster']=y_predict
```
